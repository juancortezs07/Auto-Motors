```python
# Eliminamos nulos
ventas = ventas.dropna()
```

```python
# Revisamos los tipos de datos
ventas.info()
```

```python
# Convertir  a enteros y flotantes
ventas["Empleado"] = ventas["Empleado"].astype(int)
ventas["Ventas"] = ventas["Ventas"].astype(float)
```

```python
# Se ratifica que Ventas es string
ventas['Ventas'] = ventas['Ventas'].astype(str)

# Usamos str.contains() para buscar registros con comas o puntos en la columna 'Ventas'
registros_con_comas_o_puntos = ventas[ventas['Ventas'].str.contains('[,.]')]

# Mostramos los registros encontrados y vemos que existen 5. Se ve que sobran dos ceros despuÃ©s de la coma
registros_con_comas_o_puntos
```

```python
# FunciÃ³n para eliminar los ceros a la derecha de la coma
def eliminar_ceros_despues_coma(cadena):
    if ',' in cadena:
        partes = cadena.split(',')
        parte_despues_coma = partes[1].rstrip('0')
        return partes[0] + (',' + parte_despues_coma if parte_despues_coma else '')
    return cadena

# Aplicamos la funciÃ³n y volvemos a guardar
ventas['Ventas']= ventas['Ventas'].apply(eliminar_ceros_despues_coma)
```

```python
# Utiliza el mÃ©todo loc para seleccionar los registros con los Ã­ndices 6720, 6721, 45141, 58461 y 78818 para verificar los cambios
indices_a_buscar = [6720, 6721, 45141, 58461, 78818]
registros_seleccionados = ventas.loc[indices_a_buscar]

registros_seleccionados
```
